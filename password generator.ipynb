# Auto Password Generator  using GUI

```python
from tkinter import*
import string
import pyperclip
import random


window = Tk()
window.geometry("370x300") 
pstr = StringVar()
plen = IntVar()
plen.set(0)
c=StringVar()
operator=""


# function to generate the password
def generate():
    try:
        if(plen.get()==0):
            d=("Enter A valid number")
            c.set(d)
        else:
            s1 = string.ascii_lowercase
            s2 = string.ascii_uppercase
            s3 = string.digits
            s4 = string.punctuation  
            s = []
            s.extend(list(s1))
            s.extend(list(s2))
            s.extend(list(s3))
            s.extend(list(s4))
            password=("".join(random.sample(s, plen.get())))
            pstr.set(password)
            d=("Successfully Created")
            c.set(d)
    except:
        d=("Enter A valid number")
        c.set(d)

    
    
def copytoclipboard():
    random_password = pstr.get()
    pyperclip.copy(random_password)
    
    
    
def btnClear():
    global operator
    operator = ""
    c.set("")
    pstr.set("")
    pyperclip.copy("")
    plen.set("")

# Creating a text label widget
Label(window, text="Password Generator Application", font="calibri 20 bold",bg="cyan").pack()

# Creating a text label widget
Label(window, bg="lightblue",text="Enter password length").pack(pady=3)

# Creating a entry widget to take password length entered by the 
# user
Entry(window, textvariable=plen, bg="cyan").pack(pady=3)

# button to call the generate function
Button (window, text="Generate Password",bg="lightblue" ,command=generate).pack(pady=7)

# entry widget to show the generated password
Entry(window, textvariable=c, bg="cyan").pack(pady=3)

# entry widget to show the generated password
Entry(window, textvariable=pstr, bg="cyan").pack(pady=3)

# button to call the copytoclipboard function
Button(window, text="Copy to clipboard",bg="lightblue" , command=copytoclipboard ).pack()

buttonClr = Button(window,bd=8,font="calibiri 20 bold",text="C",bg="red",command= btnClear).pack()
window.mainloop()
```
